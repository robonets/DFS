```python
# When we calculate dz/dx, we also need to derive dy/dx.
# Here, we can use double backprop

import sys
import os

cwd = os.getcwd()
parent_dir = os.path.abspath(os.path.join(cwd, os.pardir))
sys.path.append(parent_dir)

import numpy as np
from dezero import Variable

x = Variable(np.array(2.0))
y = x ** 2
y.backward(create_graph=True) # create a computational graph (2x)
gx = x.grad # this is not a value but a computational graph
x.cleargrad()

z = gx ** 3 + y
z.backward()
print(x.grad)

```
